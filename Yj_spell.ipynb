```python
import zipfile
import os

# 1. Zipファイルが含まれているディレクトリのパス
zip_files_dir = "/content/drive/MyDrive/4year/J-FingerSpell_raw/kana"  # ディレクトリのパスを指定

# 2. Zipファイルを展開するディレクトリ
extract_base_dir = "/content/drive/MyDrive/4year/J-FingerSpell_raw/raw_date_kana"  # 展開先のディレクトリ

# 3. ディレクトリ内のすべてのZipファイルを探して展開
for zip_file in os.listdir(zip_files_dir):
    if zip_file.endswith(".zip"):  # Zipファイルのみを対象
        zip_file_path = os.path.join(zip_files_dir, zip_file)  # 各Zipファイルのパスを取得
        extract_dir = os.path.join(extract_base_dir, os.path.splitext(zip_file)[0])  # Zipファイル名で展開先を指定

        # Zipファイルを展開
        with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
            print(f"Extracting {zip_file} to {extract_dir}")
            zip_ref.extractall(extract_dir)  # 指定されたディレクトリに展開

print("All zip files have been extracted.")
```

```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# 特徴量とラベルのデータフレームを読み込む（例として全てのデータを格納していると仮定）
data_path = '/content/drive/MyDrive/4year/J-FingerSpell_raw/Svm_Model/normalized_all'  # 正規化されたデータのディレクトリ
labels_path = '/content/drive/MyDrive/4year/J-FingerSpell_raw/Svm_Modellabels_svm.csv'  # 正解ラベルのファイル

# 正解ラベルを読み込む
labels_df = pd.read_csv(labels_path)

# 特徴量とラベルのデータフレームを作成する
all_features = []
all_labels = []

# 特徴量ファイルの処理
for root, dirs, files in os.walk(data_path):
    for file in files:
        if file.endswith('.csv'):
            # ファイルごとにデータを読み込む
            file_path = os.path.join(root, file)
            df = pd.read_csv(file_path)

            # 'normalized_x', 'normalized_y', 'normalized_z' を平坦化して特徴量にする
            flattened_features = df[['normalized_x', 'normalized_y', 'normalized_z']].values.flatten()

            # ファイル名からラベルを取得し、対応するラベルを取得
            label = extract_label_from_filename(file)

            # 特徴量とラベルをそれぞれリストに追加
            all_features.append(flattened_features)
            all_labels.append(label)

# リストをNumpy配列に変換
X = pd.DataFrame(all_features)  # 特徴量
y = pd.Series(all_labels)  # ラベル

# データをトレーニングセットとテストセットに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# SVMモデルを定義し、トレーニング
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)

# テストデータで予測
y_pred = svm_model.predict(X_test)

# 精度を計算し、結果を表示
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# クラスごとの評価レポートを表示
print(classification_report(y_test, y_pred))
```
